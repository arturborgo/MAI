```python
import numpy as np
import matplotlib.pyplot as plt

# Print 'Hello, world!'
print('Hello, world!')

# Generate data for a sine wave
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Plot the sine wave
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```
